```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('../scraper/data/ads_storage.db')
cursor = conn.cursor()

# Fetch all data
cursor.execute("SELECT * FROM ads_raw")
columns = [description[0] for description in cursor.description]
rows = cursor.fetchall()

# Convert to DataFrame
df = pd.DataFrame(rows, columns=columns)

# print(f"Total ads: {len(df)}")

conn.close()
df.head()
```

```python
conn = sqlite3.connect('../scraper/data/ads_storage.db')
cursor = conn.cursor()
# Fetch all data
cursor.execute("SELECT hash_id, link FROM ads_raw WHERE status = 'pending'")
columns = [description[0] for description in cursor.description]
rows = cursor.fetchall()

# Convert to DataFrame
df = pd.DataFrame(rows, columns=columns)
df.head()
```
