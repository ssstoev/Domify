```python
# 1. convert the price & size columns to float
def convert_strip_and_convert_to_numeric(df: pd.DataFrame, cols_to_strip: list)->pd.DataFrame:
    df[cols_to_strip] = df[cols_to_strip].apply(lambda row: row.str.replace(" ", ""))
    df[cols_to_strip] = df[cols_to_strip].astype(float)
    return df

cols_to_strip = ["price_m2_eur", "price_m2_bgn"]
df_copy = df.copy()
df_copy = convert_strip_and_convert_to_numeric(df_copy, cols_to_strip)
df_copy.info()
```

```python
import re 

match = re.search(r'\d+', str("11 ot 25"))
print(match.group())
```

```python
import numpy as np

# 2. Extract the floor number from the floor column
def extract_floor_number(floor_string: str):
    if pd.isna(floor_string):
        return np.nan
    
    match = re.search(r'\d+', str(floor_string))
    return int(match.group()) if match else np.nan

df_copy["floor_clean"] = df_copy["floor"].apply(extract_floor_number)
df_copy = df_copy.drop(labels="floor", axis=1)
df_copy.head()
```

```python
# 3. convert the yes/no columns to bool
def convert_to_bool(value: str):
    if pd.isna(value):
        return np.nan
    
    elif value.upper() == "ДА":
        return True

    else:
        return False

bool_cols = ["akt16", "broker_commision"]
for col in bool_cols:
    df_copy[f"{col}_processed"] = df_copy[f"{col}"].apply(convert_to_bool)
    df_copy.drop(labels=col, axis=1, inplace=True)

df_copy.head()
```
