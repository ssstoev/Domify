```python
# 1. convert the price & size columns to float
def convert_strip_and_convert_to_numeric(df: pd.DataFrame, cols_to_strip: list)->pd.DataFrame:
    df[cols_to_strip] = df[cols_to_strip].apply(lambda row: row.str.replace(" ", ""))
    df[cols_to_strip] = df[cols_to_strip].astype(float)
    return df

cols_to_strip = ["price_m2_eur", "price_m2_bgn"]
df_copy = df.copy()
df_copy = convert_strip_and_convert_to_numeric(df_copy, cols_to_strip)
df_copy.info()
```

```python
import re 

match = re.search(r'\d+', str("11 ot 25"))
print(match.group())
```

```python
import numpy as np

# 2. Extract the floor number from the floor column
def extract_floor_number(floor_string: str):
    if pd.isna(floor_string):
        return np.nan
    
    match = re.search(r'\d+', str(floor_string))
    return int(match.group()) if match else np.nan

df_copy["floor_clean"] = df_copy["floor"].apply(extract_floor_number)
df_copy = df_copy.drop(labels="floor", axis=1)
df_copy.head()
```

```python
df["broker_commision"].unique().tolist()
```

```python
# 3. convert the yes/no columns to bool
def convert_to_bool(value: str):
    if pd.isna(value):
        return np.nan
    
    elif value.upper() == "ДА":
        return True

    else:
        return False

bool_cols = ["akt16", "broker_commision"]
for col in bool_cols:
    df_copy[f"{col}_processed"] = df_copy[f"{col}"].apply(convert_to_bool)
    df_copy.drop(labels=col, axis=1, inplace=True)

df_copy.head()
```

```python
df_copy.loc[3, "link"]
```

```python
# 4. Extract the neighbourhood from the title
df_copy['neighbourhood'] = df_copy['title'].str.extract(r'в София, \s*(.*)', expand=False)
df_copy.head()
```

```python
from collections import Counter

# Collect all extras by splitting strings
all_items = []
for extras_str in df_copy['extras'].dropna():
    items = [x.strip() for x in extras_str.split(';')]
    all_items.extend(items)

# Count frequency
extra_counts = Counter(all_items)

# See most common extras
print(f"Total unique extras: {len(extra_counts)}\n")
for extra, count in extra_counts.most_common(30):
    print(f"{extra}: {count} ({count/len(df_copy)*100:.1f}%)")
```

```python
# 6. Extract if its a garage or living building or area for building (parcel)
# "Гараж, паркомясто"; "Парцел"

# we use a vectorized approach since it's fastest
df_copy["type_of_estate"] = np.select(
    [
        df_copy["title"].str.contains("Гараж", na=False),
        df_copy["title"].str.contains("Парцел", na=False),
        df_copy["title"].str.contains("Магазин", na=False),
    ],
    ["гараж", "парцел", "магазин"],
    default="жилище"
)

df_copy["type_of_estate"].value_counts()
```
