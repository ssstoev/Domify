```python
import sqlite3

conn = sqlite3.connect('../scraper/data/ads_storage.db')
cursor = conn.cursor()
# Fetch all data
cursor.execute('''
SELECT
    price_m2_eur,
    size_m2,
    COALESCE(price_m2_eur,0) * COALESCE(size_m2,0) AS computed,
    total_price_eur
FROM ads_cleaned
WHERE total_price_eur IS NULL
LIMIT 20;
               ''')
pending = cursor.fetchall()
conn.close()
pending
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('../scraper/data/ads_storage.db')
cursor = conn.cursor()

# List all tables in the database
cursor.execute("SELECT name FROM sqlite_master WHERE type='table' ORDER BY name;")
tables = cursor.fetchall()
conn.close()

print("Tables in database:")
for t in tables:
    print(" -", t[0])
```
